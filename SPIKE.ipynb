# SPIKE SESSION WITH OTTAVIA


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv("/Users/sandraakunnaejiofor/Documents/CODE ACADEMY BERLIN/MATPLOTLIB_AND_SEABORN/movies.csv")
```

```python
movies.dtypes
```

```python
movies.drop([615], axis=0, inplace=True)
```

```python
movies.drop(columns=["Summary", "MovieID"], inplace=True)
```

```python
movies["Budget"] = movies["Budget"].astype("int")
movies["Gross"] = movies["Gross"].astype("int")
movies["Release Date"] = pd.to_datetime(
    movies["Release Date"], format='%Y-%m-%d', errors='ignore')
```

```python
movies.dtypes
```

# BAR CHART / COUNTPLOT

MATPLOTLIB

```python
unique_genres_count = movies["Genre"].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(unique_genres_count.index, unique_genres_count, width=0.9, color="pink")
plt.title("title")
plt.xlabel("Genres")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# HISTOGRAM

# matplotlib

```python
plt.figure(figsize=(15, 6))
plt.title("title")
plt.hist(movies['Gross'], color='purple', bins=80,
         edgecolor='grey', linewidth=0.5)
plt.show()        
```

```python
plt.figure(figsize=(15, 6))
plt.title("title")
plt.hist(movies['Budget'], color='lightgreen', bins=80,
         edgecolor='grey', linewidth=0.5)
plt.show() 
```

```python
fig, axs = plt.subplots(1, 2, sharey="row", tight_layout=True)

axs[0].set_title("Budget")
axs[1].set_title("Gross")

axs[0].hist(movies['Budget'], color='pink', bins=80,
        edgecolor='grey', linewidth=0.5)
axs[1].hist(movies['Gross'], color='lightgreen', bins=80,
        edgecolor='grey', linewidth=0.5)
plt.show()               
```

```python
fig, axs = plt.subplots(1, 2, sharey="row", tight_layout=True)
axs[0].set_title("Budget")
axs[1].set_title("Gross")
sns.histplot(data=movies, x="Budget",
            color='purple', ax=axs[0], bins=80, edgecolor='black', linewidth=0.5)
sns.histplot(data=movies, x="Gross",
            color='yellow', ax=axs[1], bins=80, edgecolor='black', linewidth=0.5)           
```

# BOXPLOTS

# matplotlib

```python
genre_budget = {}
for i in movies['Genre'].unique():
    genre_budget[i] = movies[movies['Genre'] == i].Budget
plt.boxplot([x for x in genre_budget.values()],
            labels=[x for x in genre_budget.keys()])
#x and y-axis labels
plt.xlabel('Genre')
plt.ylabel('Budget')
plt.xticks(rotation=45)
#plot title
plt.title('Relation between Genre and Budget')
plt.show()            
```
