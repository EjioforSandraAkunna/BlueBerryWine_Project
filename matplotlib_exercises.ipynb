# BLUEBERRY WINERY PROJECT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
red_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=";")
```

```python
white_wine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", sep=";")
```

```python
unique_wine_count = white_wine["citric acid"].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(unique_wine_count.index, unique_wine_count, width=0.9, color="pink")
plt.title("CHECKINGS")
plt.xlabel("citric acid")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.xticks(rotation=45)
plt.title("CHECKINGS")
sns.countplot(x=white_wine["citric acid"], order=white_wine["citric acid"].value_counts(
).index, color="lightgreen")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.xticks(rotation=45)
plt.title("CHECKINGS")
sns.countplot(x=white_wine["citric acid"], color="lightgreen")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, sharey="row", tight_layout=True)

axs[0].set_title("free sulfur dioxide")
axs[1].set_title("total sulfur dioxide")

axs[0].hist(white_wine['free sulfur dioxide'], color='pink', bins=80,
        edgecolor='grey', linewidth=0.5)
axs[1].hist(white_wine['total sulfur dioxide'], color='lightgreen', bins=80,
        edgecolor='grey', linewidth=0.5)
plt.show() 
```

```python
fig, axs = plt.subplots(1, 2, sharey="row", tight_layout=True)

axs[0].set_title("citric acid")
axs[1].set_title("chlorides")

axs[0].hist(white_wine['citric acid'], color='pink', bins=80,
        edgecolor='grey', linewidth=0.5)
axs[1].hist(white_wine['chlorides'], color='lightgreen', bins=80,
        edgecolor='grey', linewidth=0.5)
plt.show() 
```

```python
fig, axs = plt.subplots(1, 2, sharey="row", tight_layout=True)

axs[0].set_title("fixed acidity")
axs[1].set_title("volatile acidity")

axs[0].hist(white_wine['fixed acidity'], color='pink', bins=80,
        edgecolor='grey', linewidth=0.5)
axs[1].hist(white_wine['volatile acidity'], color='lightgreen', bins=80,
        edgecolor='grey', linewidth=0.5)
plt.show() 
```

```python
fig, axs = plt.subplots(1, 2, sharey="row", tight_layout=True)
axs[0].set_title("fixed acidity")
axs[1].set_title("volatile acidity")
sns.histplot(data=white_wine, x="fixed acidity",
            color='purple', ax=axs[0], bins=80, edgecolor='black', linewidth=0.5)
sns.histplot(data=white_wine, x="volatile acidity",
            color='yellow', ax=axs[1], bins=80, edgecolor='black', linewidth=0.5) 
```

```python
fig, axs = plt.subplots(1, 2, sharey="row", tight_layout=True)
axs[0].set_title("fixed acidity")
axs[1].set_title("volatile acidity")
sns.histplot(data=red_wine, x="fixed acidity",
            color='purple', ax=axs[0], bins=80, edgecolor='black', linewidth=0.5)
sns.histplot(data=red_wine, x="volatile acidity",
            color='yellow', ax=axs[1], bins=80, edgecolor='black', linewidth=0.5) 
```

```python
wine_values = [white_wine.shape[0], red_wine.shape[0]]
wine_values
```

```python
fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')
labels = ["White wine", "Red wine"]
wine_values = [white_wine.shape[0], red_wine.shape[0]]
explode = (0, 0.1)

plt.pie(wine_values, explode=explode, labels=labels,radius=1.5, startangle=90, autopct='%1.1f%%')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, sharey="row", tight_layout=True, figsize=(10, 6))
axs[0].set_title("White Wine")
axs[1].set_title("Red Wine")
sns.histplot(data=white_wine, x="quality",
            color='purple', ax=axs[0], bins=80, edgecolor='black', linewidth=0.5)
sns.histplot(data=red_wine, x="quality",
            color='yellow', ax=axs[1], bins=80, edgecolor='black', linewidth=0.5)
```
